## Step 1 — Install Python

1. Go to https://www.python.org/downloads/ and download the `.pkg` file for Python 3.11+.
2. Run the installer and follow the prompts.
3. After installation, open Terminal and run the certificate script that appears in the Python folder.

**Option B — Homebrew** (if you already use it):
```
brew install python@3.11
```

### Linux (Ubuntu / Debian)
Python 3 is usually pre-installed. Check with `python3 --version`. If you need to install or update:
```
sudo apt update && sudo apt install -y python3 python3-venv python3-pip
```

**Fedora / RHEL:**
```
sudo dnf install python3 python3-pip
```

> **Convention:** On macOS and Linux, use `python3`. On Windows, use `py` (the Python Launcher) or `python`.

## Step 4 — Create a local virtual environment (.venv)

A virtual environment keeps this project's packages completely separate from the rest of your system. It lives inside the project folder in a directory called `.venv`.

Run the appropriate command in your terminal:

| Platform | Command |
|---|---|
| Windows | `py -m venv .venv` |
| macOS | `python3 -m venv .venv` |
| Linux | `python3 -m venv .venv` |

> **Check:** a new `.venv` folder should appear in the project directory. The `.venv` folder is already excluded from git (see `.gitignore`).

## Step 5 — Activate the virtual environment

Activation must be done in your **terminal** (it cannot be done from inside a notebook cell).

| Platform | Shell | Command |
|---|---|---|
| **Windows** | **Command Prompt** ✅ | `.venv\Scripts\activate.bat` |
| Windows | PowerShell ⚠ | `.venv\Scripts\Activate.ps1` |
| macOS | bash / zsh | `source .venv/bin/activate` |
| Linux | bash / zsh | `source .venv/bin/activate` |

After activation, your prompt changes to show `(.venv)` at the start:
```
(.venv) C:\Users\you\project>
```

---

### ⚠ Windows PowerShell — "running scripts is disabled" error

This is a common Windows security setting. **Two options:**

**Option A — Quickest fix: use Command Prompt instead of PowerShell.**
Press `Win + R`, type `cmd`, press Enter, navigate to the project folder, then run:
```
.venv\Scripts\activate.bat
```

**Option B — Fix PowerShell permanently for your account.**
Run this once in any PowerShell window (no Administrator rights needed):
```
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser -Force
```
Then retry `.venv\Scripts\Activate.ps1`.

## Step 9 — Install Ollama and download local models

We will also be using the LLM **DeepSeek**. This requires installing **Ollama** on your computer.


This downloads the model weights to your computer so they can be called locally from Python via LangChain.

````python
# Ollama + LangChain smoke test
# Run this after installing Ollama and pulling at least one model.

import os
import shutil
import subprocess
from pathlib import Path

try:
    from langchain_ollama.llms import OllamaLLM
except ImportError:
    raise ImportError(
        "langchain-ollama is not installed. Re-run Step 6: pip install -r requirements.txt"
    )


def find_ollama_executable() -> str:
    """Find Ollama CLI across Windows/macOS/Linux and notebook PATH quirks."""
    # 1) Normal PATH lookup
    path_hit = shutil.which("ollama")
    if path_hit:
        return path_hit

    # 2) WSL/Git Bash edge case on Windows host
    path_hit_exe = shutil.which("ollama.exe")
    if path_hit_exe:
        return path_hit_exe

    # 3) Common Windows install locations
    if os.name == "nt":
        candidates = [
            Path.home() / "AppData" / "Local" / "Programs" / "Ollama" / "ollama.exe",
            Path("C:/Program Files/Ollama/ollama.exe"),
        ]
        for c in candidates:
            if c.exists():
                return str(c)

    return ""


ollama_cmd = find_ollama_executable()
if not ollama_cmd:
    raise FileNotFoundError(
        "Ollama executable not found from this notebook kernel.\n"
        "Try one of these fixes:\n"
        "1) Restart VS Code completely and re-open this notebook.\n"
        "2) Confirm Ollama is installed: https://ollama.com/\n"
        "3) On Windows, verify this file exists:\n"
        "   C:/Users/<you>/AppData/Local/Programs/Ollama/ollama.exe"
    )

# 1) Check that the Ollama CLI is available
check = subprocess.run([ollama_cmd, "--version"], capture_output=True, text=True)
if check.returncode != 0:
    raise RuntimeError(
        "Ollama CLI found but failed to run. Output:\n"
        f"{check.stdout}\n{check.stderr}"
    )

print("Ollama CLI:", (check.stdout or check.stderr).strip())
print("Using executable:", ollama_cmd)

# 2) Check local models
models = subprocess.run([ollama_cmd, "list"], capture_output=True, text=True, check=True)
model_list_text = models.stdout
print("\nInstalled models:\n", model_list_text)

preferred_models = ["deepseek-r1:1.5b", "phi3:3.8b"]
selected_model = None
for m in preferred_models:
    if m in model_list_text:
        selected_model = m
        break

if selected_model is None:
    raise RuntimeError(
        "None of the expected models were found. Run:\n"
        "  ollama pull deepseek-r1:1.5b\n"
        "  ollama pull phi3:3.8b\n"
    )

print(f"\nUsing model: {selected_model}")

# 3) LangChain -> Ollama request
llm = OllamaLLM(model=selected_model, temperature=0)
reply = llm.invoke("Reply with exactly this text: OLLAMA_CONNECTION_OK")
print("\nModel reply:", reply)

if "OLLAMA_CONNECTION_OK" in reply:
    print("\nSUCCESS: LangChain can communicate with Ollama.")
else:
    print("\nWarning: communication worked, but response text was unexpected.")
````

## Troubleshooting

**`python3: command not found` / `py: command not found`**
- Windows: ensure "Add Python to PATH" was ticked during installation, then restart the terminal.
- macOS: use `python3`, not `python`. If still missing, re-run the official installer from python.org.
- Linux: run `sudo apt install python3 python3-venv python3-pip` (Debian/Ubuntu).

**PowerShell — "running scripts is disabled" (Windows)**
Two options:
1. **Easiest:** switch to Command Prompt and run `.venv\Scripts\activate.bat` instead.
2. **Fix PowerShell:** run this once in any PowerShell window:
   ```
   Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser -Force
   ```

**Kernel not showing in VS Code**
1. Press `Ctrl+Shift+P` → *Python: Select Interpreter* → pick the `.venv` entry.
2. Reload the window: `Ctrl+Shift+P` → *Developer: Reload Window*.
3. If still missing, re-run the kernel registration command (Step 7).

**Packages not found after installation**
Check the active kernel is **Python (.venv) CESAB 2026**. If Step 6 was run with the wrong environment, activate `.venv` in the terminal and re-run Step 6.

**Ollama command not found**
- Install Ollama from https://ollama.com/.
- Fully close and reopen VS Code/terminal so the `ollama` command is on PATH.
- Verify with:
  ```
  ollama --version
  ```

**Smoke test fails with missing model**
Run one of:
```
ollama pull deepseek-r1:1.5b
ollama pull phi3:3.8b
ollama pull deepseek-r1:7b
```

**`pip install` fails on Linux — missing system libraries**
```
sudo apt install build-essential libffi-dev libssl-dev
```

**macOS SSL error when downloading models**
```
/Applications/Python\ 3.x/Install\ Certificates.command
```
Replace `3.x` with your Python version number.
